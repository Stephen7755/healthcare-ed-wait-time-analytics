```python
import pandas as pd
```

```python
df = pd.read_csv("emergency_department_wait_times.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
#convert time arrival

df["arrival_time"] = pd.to_datetime(df["arrival_time"])
```

```python
##average ED wait time 

df["wait_time_minutes"].mean()
```

```python
##wait time by triage

df.groupby("triage_level")["wait_time_minutes"].mean().sort_values(ascending=False)
```

```python
##wait time outcome

df.groupby("outcome")["wait_time_minutes"].mean()
```

```python
#Wait Time by Hour of Day (Very Real-World)
df["arrival_hour"] = df["arrival_time"].dt.hour
```

```python

df.groupby("arrival_hour")["wait_time_minutes"].mean()
```

```python
df.groupby("triage_level")["wait_time_minutes"].mean().sort_values(ascending=False)
```

```python
df["arrival_hour"] = df["arrival_time"].dt.hour
df.groupby("arrival_hour")["wait_time_minutes"].mean()
```

```python
df["arrival_hour"] = df["arrival_time"].dt.hour
df["arrival_day"] = df["arrival_time"].dt.day_name()

df.to_csv("cleaned_emergency_department_wait_times.csv", index=False)
```
